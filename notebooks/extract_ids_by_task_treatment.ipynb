# Extract `experimental_metadata.id` lists by Task/Treatment

This notebook queries the Postgres database `dlc_pose_analysis_2026` and returns two ID lists:
- Task = ChickenBroth, Treatment = Y
- Task = ChickenBroth, Treatment = P

If your `task` values use a different spelling/case, run the distinct-values cell and update the `TASK` variable.

```python
import psycopg2

DB_CONNECT_KWARGS = {
    "host": "localhost",
    "port": 5432,
    "user": "atanugiri",
    "password": "",
    "database": "dlc_pose_analysis_2026",
}

def fetch_ids(task: str, treatment: str):
    sql = """
    SELECT id
    FROM public.experimental_metadata
    WHERE task = %s AND treatment = %s
    ORDER BY id
    """
    with psycopg2.connect(**DB_CONNECT_KWARGS) as conn:
        with conn.cursor() as cur:
            cur.execute(sql, (task, treatment))
            return [row[0] for row in cur.fetchall()]

ids_y = fetch_ids("ChickenBroth", "Y")
ids_p = fetch_ids("ChickenBroth", "P")
print("Y:", ids_y)
print("P:", ids_p)
```
